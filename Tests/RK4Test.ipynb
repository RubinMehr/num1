```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def f(t, y):
    return y

def runge_kutta(y0: float, t0: float, T: float, h: float) -> tuple:
    n = int(T/h)
    ts = np.linspace(t0, T, n)
    y = np.zeros(n)
    y[0] = y0

    for i in range(0,n-1):
        k1 = f(t0, y[i])
        k2 = f(t0, y[i]+(h*k1/2))
        k3 = f(t0, y[i]+(h*k2/2))
        k4 = f(t0, y[i]+(h*k3))
        y[i+1] = y[i] + h/6 * (k1+2*k2+2*k3+k4)

    return ts, y

h = 2**-5
t0 = 0
y0 = 1
T = 5

ts, y = runge_kutta(y0, t0, T, h)
sol = scipy.integrate.solve_ivp(f,[t0,T],[y0])

plt.figure(figsize=(10, 6))
plt.plot(ts, y, label="Meine Trajektorie")
#plt.plot(sol['t'], sol['y'][0], label="Trajektorie")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title("Runge-Kutta Methode")
plt.show()
```
