In diesem Übungsblatt wird im Prinzip nur geplottet. Beim ersten Plot ist $u_1(t)$ wie die numerische Lösung $= 0$, von daher scheint es so als würde sie nicht geplottet werden, allerdings ist sie tatsächlich in der Graphik enthalten.

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def f(t, u):
    return np.sqrt(np.sqrt(u))

def u2(t):
    return np.cbrt(3/4 * t)**4

sol = scipy.integrate.solve_ivp(f, [0,2], [0])
t = sol['t']
y = sol['y'][0]

zeroes = np.zeros(t.shape[0])

# Uninteressantes Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, y, label="Numerische Lösung")
plt.plot(t, zeroes, label='$u_1(t)$')
plt.plot(t, u2(t), label = '$u_2(t)$')
plt.xlabel("t")
plt.ylabel("u(t)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title("Lösung für die nichtgestörte ODE")
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(t, y, label="Numerische Lösung")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title("Numerische Lösung für die nichtgestörte ODE")
plt.plot()
```

```python
eps = 0

def f(t,u):
    global eps
    return np.sqrt(np.sqrt(u + eps))

for k in range(1,6):
    eps = 10**-k
    sol = scipy.integrate.solve_ivp(f, [0,2], [0])
    t = sol['t']
    y = sol['y'][0]

    zeroes = np.zeros(t.shape[0])

    # Uninteressantes Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(t, y, label="$v_\epsilon(t)$")
    plt.plot(t, u2(t), label = '$u_2(t)$')
    plt.xlabel("t")
    plt.ylabel("u(t)")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.title("Lösung für die gestörte ODE mit $\epsilon$ = " + f'{eps}')
    plt.show()
```
