```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-5,5,300)
b = np.linspace(-5,5,300)
```

```python
plot = np.array([])
for i in range(0, a.shape[0]):
    for j in range(0, a.shape[0]):
        z = complex(a[i], b[j])
        if np.abs(z+1) <= 1:
            plot = np.append(plot, z)

plt.figure(figsize=(10, 10))
plt.plot(np.real(plot[:]), np.imag(plot[:]), marker='.', label='$z$', linestyle='none')
plt.xlabel("Re($z$)")
plt.ylabel("Im($z$)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title(f"Stabilitätsgebiet expl. Euler")
plt.show()
```

```python
plot = np.array([])
for i in range(0, a.shape[0]):
    for j in range(0, a.shape[0]):
        z = complex(a[i], b[j])
        if np.abs(1/(1-z)) <= 1:
            plot = np.append(plot, z)

plt.figure(figsize=(10, 10))
plt.plot(np.real(plot[:]), np.imag(plot[:]), marker='.', label='$z$', linestyle='none')
plt.xlabel("Re($z$)")
plt.ylabel("Im($z$)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title(f"Stabilitätsgebiet impl. Euler")
plt.show()
```

```python
plot = np.array([])
for i in range(0, a.shape[0]):
    for j in range(0, a.shape[0]):
        z = complex(a[i], b[j])
        if np.abs((2+z)/(2-z)) <= 1:
            plot = np.append(plot, z)

plt.figure(figsize=(10, 10))
plt.plot(np.real(plot[:]), np.imag(plot[:]), marker='.', label='$z$', linestyle='none')
plt.xlabel("Re($z$)")
plt.ylabel("Im($z$)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title(f"Stabilitätsgebiet Trapezregel")
plt.show()
```

```python
plot = np.array([])
for i in range(0, a.shape[0]):
    for j in range(0, a.shape[0]):
        z = complex(a[i], b[j])
        w = 1 + z + (1/2) * z**2 + (1/6) * z**3 + (1/24) * z**4
        if np.abs(w) <= 1:
            plot = np.append(plot, z)

plt.figure(figsize=(10, 10))
plt.plot(np.real(plot[:]), np.imag(plot[:]), marker='.', label='$z$', linestyle='none')
plt.xlabel("Re($z$)")
plt.ylabel("Im($z$)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title(f"Stabilitätsgebiet RK4")
plt.show()
```

Offensichtlich sind das expl. Eulerverfahren und RK4 beschränkt, das impl. Eulerverfahren und die Trapezregel sind unbeschränkt. Weiterhin sind nur das impl. Eulerverfahren und die Trapezregel A-stabil, da bei beide Verfahren $\{z \in \mathbb{C}: Re(z) \leq 0\}$ (den gesamten Imaginärteil) als Teilmenge ihres Stabilitätsgebiet haben.
